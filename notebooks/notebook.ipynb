# Coffee Shop Sales Analysis


## Problem
Identify peak hours, weekday/weekend differences, seasonal trends, popular drinks, spending patterns, and price elasticity to support data-driven decisions for a coffee shop manager.

## Dataset
Daily Coffee Transactions (Kaggle, accessed 08 April 2026)

## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import mannwhitneyu
```

```python
# Create output folders and define a helper for saving figures
from pathlib import Path

FIGURES_DIR = Path('../figures') if Path('../figures').exists() or Path.cwd().name == 'notebooks' else Path('figures')
FIGURES_DIR.mkdir(exist_ok=True)

def save_current_figure(filename):
    plt.tight_layout()
    plt.savefig(FIGURES_DIR / filename, dpi=300, bbox_inches='tight')
```

```python
# Drop rows with missing critical values (if any)
df = df.dropna(subset=['money', 'coffee_name', 'cash_type'])
```

## 5. Basic Descriptive Analysis

```python
# Total transactions, average price, etc.
total_transactions = len(df)
avg_price = df['money'].mean()
print(f"Total transactions: {total_transactions}")
print(f"Average transaction value: ${avg_price:.2f}")
```

```python
# Most popular drinks
top_drinks = df['coffee_name'].value_counts().head(5)
print("\nTop 5 most popular drinks:")
print(top_drinks)
```

## 6. Time-based Analysis

```python
# 6.1 Hourly sales volume
hourly_sales = df.groupby('hour').size().reset_index(name='count')
peak_hour = hourly_sales.loc[hourly_sales['count'].idxmax(), 'hour']
peak_qty = hourly_sales['count'].max()
print(f"Peak hour: {peak_hour}:00 with {peak_qty} transactions")
```

```python
# Plot
plt.figure(figsize=(12,5))
plt.bar(hourly_sales['hour'], hourly_sales['count'], color='skyblue')
plt.title('Number of Transactions by Hour of Day')
plt.xlabel('Hour (24h)')
plt.ylabel('Number of Transactions')
plt.xticks(range(0,24,2))
plt.grid(axis='y', linestyle='--', alpha=0.7)
save_current_figure('hourly_transactions.png')
plt.show()
```

```python
# 6.2 Weekday vs Weekend
print("Unique values in 'weekday' column:", df['weekday'].unique())

weekend_set = {'Sat', 'Sun', 'Saturday', 'Sunday'}

if 'Weekdaysort' in df.columns:
    df['day_type'] = df['Weekdaysort'].apply(lambda x: 'Weekend' if int(x) >= 6 else 'Weekday')
else:
    df['day_type'] = df['weekday'].apply(lambda x: 'Weekend' if x in weekend_set else 'Weekday')

weekday_weekend = df.groupby('day_type').size().reindex(['Weekday', 'Weekend'])
weekday_qty = int(weekday_weekend['Weekday'])
weekend_qty = int(weekday_weekend['Weekend'])
weekday_ratio = weekday_qty / weekend_qty

print(f"Weekday transactions: {weekday_qty}, Weekend: {weekend_qty}")
print(f"Weekday is {weekday_ratio:.1f} times busier than weekend")
```

```python
# 6.3 Monthly trend
month_name_map = {
    'Jan': 'January', 'Feb': 'February', 'Mar': 'March', 'Apr': 'April',
    'May': 'May', 'Jun': 'June', 'Jul': 'July', 'Aug': 'August',
    'Sep': 'September', 'Oct': 'October', 'Nov': 'November', 'Dec': 'December'
}

if {'month', 'Monthsort'}.issubset(df.columns):
    monthly_sales = (
        df.groupby(['Monthsort', 'month'])
        .size()
        .reset_index(name='count')
        .sort_values('Monthsort')
    )
    monthly_sales['month_display'] = monthly_sales['month'].map(month_name_map).fillna(monthly_sales['month'])
    peak_row = monthly_sales.loc[monthly_sales['count'].idxmax()]
    peak_month = peak_row['month']
    peak_month_display = peak_row['month_display']
    monthly_max = int(peak_row['count'])
    print(f"Peak month: {peak_month_display} with {monthly_max} transactions")

    plt.figure(figsize=(10,5))
    plt.plot(monthly_sales['month_display'], monthly_sales['count'], marker='o', linestyle='-', color='green')
    plt.title('Monthly Transaction Trend')
    plt.xlabel('Month')
    plt.ylabel('Number of Transactions')
    plt.grid(True)
    save_current_figure('monthly_transaction_trend.png')
    plt.show()
```

## 7. Behavioral Economics Extensions

```python
# 7.1 Average spending by hour
hourly_avg_price = df.groupby('hour', as_index=False)['money'].mean().sort_values('hour')
peak_avg_row = hourly_avg_price.loc[hourly_avg_price['money'].idxmax()]
peak_avg_hour = int(peak_avg_row['hour'])
peak_avg_value = float(peak_avg_row['money'])

print("Average spending by hour:")
print(hourly_avg_price)
print(f"Highest average spending occurs at {peak_avg_hour}:00 (${peak_avg_value:.2f}).")

plt.figure(figsize=(12,5))
plt.plot(hourly_avg_price['hour'], hourly_avg_price['money'], marker='o', linestyle='-', color='orange')
plt.title('Average Transaction Value by Hour')
plt.xlabel('Hour')
plt.ylabel('Average Amount ($)')
plt.grid(True)
save_current_figure('average_spending_by_hour.png')
plt.show()
```

```python
# 7.2 Weekend vs Weekday average spending
weekday_avg_price = df.groupby('day_type')['money'].mean().reindex(['Weekday', 'Weekend'])
spending_gap = weekday_avg_price['Weekday'] - weekday_avg_price['Weekend']

if abs(spending_gap) < 0.2:
    spending_message = 'Average spending is very similar on weekdays and weekends.'
elif spending_gap > 0:
    spending_message = 'Average spending is slightly higher on weekdays.'
else:
    spending_message = 'Average spending is slightly higher on weekends.'

print(f"Weekday avg spending: ${weekday_avg_price['Weekday']:.2f}, Weekend avg: ${weekday_avg_price['Weekend']:.2f}")
print(spending_message)
```

```python
# 7.3 Sales volume by price range
price_bins = [15, 20, 30, 50]
price_labels = ['15-20', '20-30', '30-50']
df['price_range'] = pd.cut(df['money'], bins=price_bins, labels=price_labels, right=False)
price_range_sales = df['price_range'].value_counts().sort_index()
top_price_range = price_range_sales.idxmax()

print("Sales volume by price range:")
print(price_range_sales)
print(f"The busiest price range is {top_price_range}.")

plt.figure(figsize=(8,5))
price_range_sales.plot(kind='bar', color='lightcoral')
plt.title('Demand by Price Range')
plt.xlabel('Price Range ($)')
plt.ylabel('Number of Transactions')
plt.grid(axis='y', linestyle='--', alpha=0.7)
save_current_figure('price_range_demand.png')
plt.show()
```

```python
# 7.4 Top drinks: price vs popularity
top5_drinks = top_drinks.index.tolist()
drink_stats = df[df['coffee_name'].isin(top5_drinks)].groupby('coffee_name').agg(
    avg_price=('money', 'mean'),
    total_sales=('coffee_name', 'count')
).sort_values('total_sales', ascending=False)
print("Top drinks - price and sales:")
print(drink_stats)
```

```python
# Scatter plot: price vs popularity
plt.figure(figsize=(10,6))
sns.scatterplot(data=drink_stats, x='avg_price', y='total_sales', size='total_sales', sizes=(100,500), legend=False)
for i, row in drink_stats.iterrows():
    plt.text(row['avg_price']+0.5, row['total_sales']+5, i, fontsize=9)
plt.title('Drink Price vs Popularity')
plt.xlabel('Average Price ($)')
plt.ylabel('Total Sales')
plt.grid(True, linestyle='--', alpha=0.5)
save_current_figure('drink_price_vs_popularity.png')
plt.show()
```

## 8. Weekday vs Weekend Statistical Comparison

This section tests whether the observed weekday and weekend difference in transaction values is statistically meaningful rather than a result of random variation. A Mann-Whitney U test is used because it does not require the transaction-value distribution to be normally distributed.

The purpose of this step is to give staffing and inventory recommendations a stronger analytical basis while keeping the interpretation cautious: statistical significance can support confidence in an observed difference, but it does not establish causality.

```python
# 8.1 Statistical comparison of weekday vs weekend transaction values
weekday_spending = df.loc[df['day_type'] == 'Weekday', 'money']
weekend_spending = df.loc[df['day_type'] == 'Weekend', 'money']

weekday_stats = {
    'sample_size': int(weekday_spending.shape[0]),
    'mean': float(weekday_spending.mean()),
    'median': float(weekday_spending.median())
}
weekend_stats = {
    'sample_size': int(weekend_spending.shape[0]),
    'mean': float(weekend_spending.mean()),
    'median': float(weekend_spending.median())
}

mw_stat, mw_pvalue = mannwhitneyu(weekday_spending, weekend_spending, alternative='two-sided')
statistical_interpretation = (
    'The weekday and weekend transaction-value distributions are statistically different.'
    if mw_pvalue < 0.05
    else 'The weekday and weekend transaction-value distributions are not statistically different at the 5% level.'
)

weekday_weekend_stats = pd.DataFrame([
    {'day_type': 'Weekday', **weekday_stats},
    {'day_type': 'Weekend', **weekend_stats}
])

print('Weekday vs Weekend Statistical Comparison')
print(weekday_weekend_stats)
print(f"Mann-Whitney U statistic: {mw_stat:.2f}")
print(f"p-value: {mw_pvalue:.4f}")
print(statistical_interpretation)

plot_df = df[['day_type', 'money']].copy()
plt.figure(figsize=(9, 5))
sns.boxplot(
    data=plot_df,
    x='day_type',
    y='money',
    hue='day_type',
    order=['Weekday', 'Weekend'],
    hue_order=['Weekday', 'Weekend'],
    palette=['#4C72B0', '#DD8452'],
    legend=False
)
plt.title('Weekday vs Weekend Transaction Values')
plt.xlabel('Day Type')
plt.ylabel('Transaction Value ($)')
plt.grid(axis='y', linestyle='--', alpha=0.4)
save_current_figure('weekday_weekend_statistical_comparison.png')
plt.show()
```

## 9. Product Segmentation by Frequency and Revenue

Popularity rankings show which drinks sell often, but they do not distinguish between products that mainly drive traffic and products that also contribute strongly to revenue. This section segments products using both transaction frequency and total revenue so that menu placement, inventory planning, and promotion design can be based on a clearer commercial view.

Median split lines are used to create four managerial segments: high frequency and high revenue, high frequency and low revenue, low frequency and high revenue, and low frequency and low revenue.

```python
# 9.1 Product segmentation by frequency and revenue
product_segmentation = (
    df.groupby('coffee_name', as_index=False)
    .agg(
        transaction_frequency=('coffee_name', 'size'),
        total_revenue=('money', 'sum'),
        avg_revenue_per_transaction=('money', 'mean')
    )
    .sort_values(['total_revenue', 'transaction_frequency'], ascending=False)
)

frequency_median = product_segmentation['transaction_frequency'].median()
revenue_median = product_segmentation['total_revenue'].median()

segment_order = [
    'High frequency + high revenue',
    'High frequency + low revenue',
    'Low frequency + high revenue',
    'Low frequency + low revenue'
]


def assign_segment(row):
    frequency_label = 'High frequency' if row['transaction_frequency'] >= frequency_median else 'Low frequency'
    revenue_label = 'high revenue' if row['total_revenue'] >= revenue_median else 'low revenue'
    return f'{frequency_label} + {revenue_label}'


product_segmentation['segment'] = product_segmentation.apply(assign_segment, axis=1)
product_segmentation['segment'] = pd.Categorical(product_segmentation['segment'], categories=segment_order, ordered=True)
product_segmentation = product_segmentation.sort_values(['segment', 'total_revenue'], ascending=[True, False])

segment_counts = (
    product_segmentation['segment']
    .value_counts(dropna=False)
    .reindex(segment_order, fill_value=0)
    .rename_axis('segment')
    .reset_index(name='product_count')
)

print('Product segmentation summary:')
print(product_segmentation[['coffee_name', 'transaction_frequency', 'total_revenue', 'avg_revenue_per_transaction', 'segment']])
print('\nProducts per segment:')
print(segment_counts)

label_products = product_segmentation.nlargest(5, 'total_revenue')

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=product_segmentation,
    x='transaction_frequency',
    y='total_revenue',
    hue='segment',
    hue_order=segment_order,
    palette=['#2E8B57', '#7FB3D5', '#F5B041', '#C0392B'],
    s=120
)
plt.axvline(frequency_median, linestyle='--', color='grey', alpha=0.7)
plt.axhline(revenue_median, linestyle='--', color='grey', alpha=0.7)
for _, row in label_products.iterrows():
    plt.text(row['transaction_frequency'] + 5, row['total_revenue'] + 120, row['coffee_name'], fontsize=9)
plt.title('Product Segmentation by Frequency and Revenue')
plt.xlabel('Transaction Frequency')
plt.ylabel('Total Revenue ($)')
plt.legend(title='Segment', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.3)
save_current_figure('product_segmentation_frequency_revenue.png')
plt.show()
```
